```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('canada_income.csv')
df.shape
```

```python
plt.ylabel("income")
plt.xlabel("year")
plt.scatter(df.year,df.income, color="red", marker="+")
```

```python
plt.ylabel("income")
plt.xlabel("year")
plt.plot(df.year,df.income, color="red", marker="+")
```

```python
new_df = df[['year','income']]
```

```python
x = new_df['year']
y = new_df['income']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=13)
```

```python
x_test, y_test
```

```python
x_train = np.array(x_train).reshape(-1,1)
y_train = np.array(y_train).reshape(-1,1)
```

```python
reg = LinearRegression()
```

```python
reg.fit(x_train, y_train)
```

```python
intercept = reg.intercept_
intercept
```

```python
coef = reg.coef_
coef
```

```python
x_test = np.array(x_test).reshape(-1,1)
y_test = np.array(y_test).reshape(-1,1)
```

```python
x_pred_train = coef*x_test+intercept
x_pred_train.flatten()
```

```python
x_train_predicted = reg.predict(x_train)
x_train_predicted.flatten()
```

```python
plt.scatter(x_train, y_train,marker="+")
plt.plot(x_train, x_train_predicted,color='red')
plt.xlabel('year')
plt.ylabel('income')
plt.title('best fit line')
```

```python
x_test_predicted = reg.predict(x_test)
x_test_predicted.flatten()
```

```python
plt.scatter(x_test, y_test,marker="+")
plt.plot(x_test, x_test_predicted,color='red')
plt.xlabel('year')
plt.ylabel('income')
plt.title('best fit line')
```
